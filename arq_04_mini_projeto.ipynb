```python
import pandas as pd
import numpy as np
df_2024_1 = pd.read_csv('Preços semestrais - AUTOMOTIVOS_2024.01.csv', sep=';', decimal=',')
df_2024_2 = pd.read_csv('Preços semestrais - AUTOMOTIVOS_2024.02.csv', sep=';', decimal=',')
df_2025_1 = pd.read_csv('Preços semestrais - AUTOMOTIVOS_2025.01.csv', sep=';', decimal=',')
df_2025_2 = pd.read_csv('Preços semestrais - AUTOMOTIVOS_2025.02.csv', sep=';', decimal=',')
```

```python
df = pd.concat([df_2024_1, df_2024_2,df_2025_1, df_2025_2],ignore_index=True)
print(df.shape)   
display(df.head(5))
display(df.tail(5))
```

```python
#df_rj = df.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()

df_rj = df.loc[df['Estado - Sigla'] == 'RJ']

df_media_municipio_rj = df_rj.groupby('Municipio')['Valor de Venda'].mean().reset_index()

df_media_municipio_rj.sort_values(by='Valor de Venda', ascending=False, inplace=True)

df_media_municipio_rj
```
