```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_2024_1 = pd.read_csv('Preços semestrais - AUTOMOTIVOS_2024.01.csv', sep=';', decimal=',')
df_2024_2 = pd.read_csv('Preços semestrais - AUTOMOTIVOS_2024.02.csv', sep=';', decimal=',')
df_2025_1 = pd.read_csv('Preços semestrais - AUTOMOTIVOS_2025.01.csv', sep=';', decimal=',')
df_2025_2 = pd.read_csv('Preços semestrais - AUTOMOTIVOS_2025.02.csv', sep=';', decimal=',')
```

```python
df = pd.concat([df_2024_1, df_2024_2,df_2025_1, df_2025_2],ignore_index=True)
print(df.shape)   
display(df.head(5))
display(df.tail(5))
```

```python
df_produto = df.loc[df['Produto'] == 'GASOLINA']
df_produto
```

```python
#df_rj = df.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()

df_rj = df_produto.loc[df['Estado - Sigla'] == 'RJ']

df_rj.sort_values(by='Valor de Venda', ascending=False, inplace=True)

df_rj
```

```python
df_media_municipio_rj = df_rj.groupby('Municipio')['Valor de Venda'].mean().reset_index()

df_media_municipio_rj.sort_values(by='Valor de Venda', ascending=False, inplace=True)

display(df_media_municipio_rj.tail(5))
display(df_media_municipio_rj.head(5))
```

```python
plt.figure(figsize=(10, 10))
plt.barh(df_media_municipio_rj['Municipio'], df_media_municipio_rj['Valor de Venda'], color='steelblue') #cor hexadecimal buscar códigos na internet
plt.title('RJ - Municipios x Preço Médio da Gasolina')
plt.tight_layout()    
plt.show()
```

```python
array_media_municipio_rj = np.array(df_media_municipio_rj['Valor de Venda']) #array trabalha com calculos em grande volume é uma boa pratica utiliza-lo
q1 = np.percentile(array_media_municipio_rj, 25)
q2 = np.percentile(array_media_municipio_rj, 50)
q3 = np.percentile(array_media_municipio_rj, 75)

iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)
print(f"Limite superior: {limite_superior}")
print(f"Limite inferior: {limite_inferior}")
```

```python
df_media_municipio_rj_outliers = df_media_municipio_rj.loc[df_media_municipio_rj['Valor de Venda'] > limite_superior]
df_media_municipio_rj_outliers = df_media_municipio_rj.loc[df_media_municipio_rj['Valor de Venda'] < limite_inferior]
df_media_municipio_rj_outliers
```

Não existem postos praticando preços fora do padrão, no estado do Rio de Janeiro. 
Nesse contexto se o limite inferior fosse negativo, indicaria que não há outliers considerando o limite inferior.

```python
df_rj_sp_etanol = df.loc[(df['Estado - Sigla'].isin(['RJ','SP'])) & (df['Produto'] == 'ETANOL')] #isin = dois filtros na mesma coluna / & = utilizado para filtrar mais de uma coluna 
```

```python
df_media_eta_rjsp = df_rj_sp_etanol.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()

df_media_eta_rjsp.sort_values(by='Valor de Venda', ascending=False, inplace=True)

df_media_eta_rjsp
```

```python
plt.figure(figsize=(3, 3))
plt.bar(df_media_eta_rjsp['Estado - Sigla'], df_media_eta_rjsp['Valor de Venda'], color='steelblue') #cor hexadecimal buscar códigos na internet
plt.title('Etanol - Preço Médio RJ e SP')
plt.tight_layout()    
plt.show()
```

```python
df_capital_rj = df_produto.loc[df_produto['Municipio'] == 'RIO DE JANEIRO']

df_capital_rj.sort_values(by='Valor de Venda', ascending=False, inplace=True)

df_capital_rj

```

```python
df_media_capital_rj = df_capital_rj.groupby('Bairro')['Valor de Venda'].mean().reset_index()

df_media_capital_rj.sort_values(by='Valor de Venda', ascending=False, inplace=True)

display(df_media_capital_rj.tail(5))
display(df_media_capital_rj.head(5))
df_media_capital_rj
```

```python
array_media_capital_rj = np.array(df_media_capital_rj['Valor de Venda']) #array trabalha com calculos em grande volume é uma boa pratica utiliza-lo
q1 = np.percentile(array_media_capital_rj, 25)
q2 = np.percentile(array_media_capital_rj, 50)
q3 = np.percentile(array_media_capital_rj, 75)

iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)
print(f"Limite superior: {limite_superior}")
print(f"Limite inferior: {limite_inferior}")
```

```python
df_media_capital_rj_outliers = df_media_capital_rj.loc[df_media_capital_rj['Valor de Venda'] > limite_superior]

df_media_capital_rj_outliers
```

# Preço mais alto (outlier superior): Flamengo é um bairro da Zona Sul, onde a população possui uma renda maior o que contribui para preços mais elevados.
# Bairros com preços mais baixos: Anil, Jardim Sulacap, Méier e Madureira apresentam preços menores, possivelmente por terem maior quantidade de postos e consumidores mais sensíveis ao preço.
# A análise indica que os preços da Gasolina C acompanham as diferenças socioeconômicas da cidade: bairros da Zona Sul tendem a ter preços mais altos, enquanto bairros da Zona Norte e Oeste costumam registrar preços mais baixos, influenciados pela renda da população e possivelmente por custos operacionais.
