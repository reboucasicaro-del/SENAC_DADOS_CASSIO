```python
df_furto_celular_cisp = df.groupby('cisp')['roubo_em_coletivo'].sum().reset_index() #reset_index transforma em coluna

df_furto_celular_cisp = df_furto_celular_cisp.sort_values(by='roubo_em_coletivo', ascending=False)

df_furto_celular_cisp
```

```python
df_aaapai_cisp = df.groupby('cisp')['aaapai'].sum().reset_index()

df_aaapai_cisp = df_aaapai_cisp.sort_values(by='aaapai', ascending=False)

df_aaapai_cisp

#ctrl+D para substituir palavras iguais
```

```python
# Usando .loc
df_capital = df.loc[df['munic'] == 'Rio de Janeiro']

# Usando .query() — sintaxe alternativa mais legível
df_capital = df.query("munic == 'Rio de Janeiro'")

# Repetindo o groupby apenas para a capital
df_cisp_capital = df_capital.groupby('cisp')['furto_celular'].sum().reset_index()

df_cisp_capital = df_cisp_capital.sort_values(by='furto_celular', ascending=False)

df_cisp_capital
```

```python
# df_cisp_capital já foi criado no passo anterior (groupby + reset_index)
array_furto = np.array(df_cisp_capital['furto_celular'])
array_furto

#grande volume de dados trabalhar com numpay
```

```python
maximo = np.max(array_furto)
minimo = np.min(array_furto)
media = np.mean(array_furto)
mediana = np.median(array_furto)
q1 = np.percentile(array_furto, 25)
q2 = np.percentile(array_furto, 50)
q3 = np.percentile(array_furto, 75)
distancia = ((media - mediana) / mediana)


print(f'O valor máixmo é {maximo}')
print(f'O valor mínimo é {minimo}')
print(f'A Média é {media:.2f}')
print(f'A Mediana é {mediana}')
print('Q1:', q1)
print('Q2:', q2)
print('Q3:', q3)
print(f'A distância é de: {distancia:.2%}')
```

```python
# Usando .loc
df_top25 = df_cisp_capital.loc[df_cisp_capital['furto_celular'] > q3]

# Usando .query()
#df_top25 = df_cisp_capital.query('furto_celular > @q3')
print(df_top25)
```

```python
# Usando .loc
df_topmenos25 = df_cisp_capital.loc[df_cisp_capital['furto_celular'] < q1]

# Usando .query()
#df_topmenos25 = df_cisp_capital.query('furto_celular < @q3')
print(df_topmenos25)

df_topmenos25 = df_topmenos25.sort_values(by='furto_celular')

df_topmenos25
```

```python
df_roubo_veiculo_cisp = df.groupby('cisp')['roubo_veiculo'].sum().reset_index() #reset_index transforma em coluna

df_roubo_veiculo_cisp = df_roubo_veiculo_cisp.sort_values(by='roubo_veiculo', ascending=False)

df_roubo_veiculo_cisp
```

```python
array_roubo_veiculo = np.array(df_roubo_veiculo_cisp['roubo_veiculo'])
array_roubo_veiculo
```

```python
maximo = np.max(array_roubo_veiculo)
minimo = np.min(array_roubo_veiculo)
media = np.mean(array_roubo_veiculo)
mediana = np.median(array_roubo_veiculo)
q1 = np.percentile(array_roubo_veiculo, 25)
q2 = np.percentile(array_roubo_veiculo, 50)
q3 = np.percentile(array_roubo_veiculo, 75)


print(f'O valor máixmo é {maximo}')
print(f'O valor mínimo é {minimo}')
print(f'A Média é {media}')
print(f'A Mediana é {mediana}')
print('Q1:', q1)
print('Q2:', q2)
print('Q3:', q3)
```
